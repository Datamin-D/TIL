```
#1 importing libraries
import pandas as pd
from pycaret.classification import *
```

```
df = pd.read_csv('/gdrive/MyDrive/csv_for_practice/train.csv')
df
```

```
#2. setup 
exp1 = setup(df, target = 'Survived',
            ignore_features=['PassengerId'])
exp1
```

```
#3. 다양한 모델 성능 비교
compare_models()
```

```
#4. 모델생성
catboost = create_model('catboost')
```

```
# 5. 모델 튜닝
tuned_catboost = tune_model(catboost) 
#코드 바뀌어서 tuned_모델 = tune_model('모델')이 아니라 tuned_모델 = tune_model(모델 객체)로 객체 바로 써주면 됨.
```

```
#6. AUC, confusion matrix
plot_model(estimator = tuned_catboost, plot = 'auc')
```

```
plot_model(estimator = tuned_catboost, plot = 'confusion_matrix')
```

```
#7. Feature Importance

plot_model(estimator = tuned_catboost, plot = 'feature')
```

```
#8. Evaluation
evaluate_model(tuned_catboost)
```

```
#9. Prediction
test = pd.read_csv('/gdrive/MyDrive/csv_for_practice/test.csv')
tuned_catboost_final = finalize_model(tuned_catboost)
predict_model(tuned_catboost_final,test)
```
