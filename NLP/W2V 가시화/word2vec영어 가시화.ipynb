# Week 10-8 특정 영역의 Word2Vec 만들기 Lab

1. thinq_review_result.csv 파일을 열어서 review_df에 저장하시오.

```python
# 데이터 읽기

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

os.chdir(r'C:\Users\user\Desktop\lg전자 교육\10주\Week 10 실습 데이터')
review_df=pd.read_csv("thinq_review_result.csv")

review_df
```

2. 문장별 단어들의 리스트인 comment_n 컬럼을 이용해서, 전체 리뷰에서 문장별 단어 리스트를 원소로 가지는 리스트 total_sentence를 만드시오.

```python
review_df['comment_n']=review_df['comment_n'].apply(lambda x: eval(x))
review_df
```

```python
total_sentence=[]
for i in range(0,review_df.shape[0]):
    total_sentence.extend(review_df.loc[i,'comment_n'])
total_sentence
```

```python
len(total_sentence)
```

```python
total_sentence[4]
```

3. total_sentence를 사용하여 Word2Vec 모델을 만드시오.

```python
# Word2Vec 훈련하기

from gensim.models import Word2Vec
from gensim.models import KeyedVectors

model = Word2Vec(sentences=total_sentence, window=5, min_count=1, workers=4, sg=0)

# sg=0 -> CBOW, sg=1 -> Skip-gram
```

4. 만들어진 Word2Vec 모델의 크기를 확인하시오.

```python
# 완성된 임베딩 매트릭스의 크기 확인
model.wv.vectors.shape
```

5. 만들어진 Word2Vec 모델에서 "엘지"와 유사한 단어들을 찾으시오.

```python
model_result = model.wv.most_similar("엘지")
print(model_result)
```

6. 생성된 Word2Vec 모델을 thing_w2v 파일에 저장하시오.

```python
# Word2Vec Model 저장과 로드
import os
os.chdir(r'C:\Users\user\Desktop\lg전자 교육\10주\Week 10 실습 데이터')
model.wv.save_word2vec_format('thinq_w2v') # 모델 저장
#loaded_model = KeyedVectors.load_word2vec_format("thinq_e_w2v") # 모델 로드
```
