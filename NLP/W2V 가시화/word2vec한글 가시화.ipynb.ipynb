```python
# Îç∞Ïù¥ÌÑ∞ ÏùΩÍ∏∞

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

os.chdir(r'C:\Users\user\Desktop\lgÏ†ÑÏûê ÍµêÏú°\10Ï£º\Week 10 Ïã§Ïäµ Îç∞Ïù¥ÌÑ∞')
review_df=pd.read_csv("thinq_e_review_result.csv")

review_df
```

```python
review_df['comment_n']=review_df['comment_n'].apply(lambda x: eval(x))
review_df
```

```python
# Word2Vec ÌõàÎ†®ÌïòÍ∏∞

from gensim.models import Word2Vec
from gensim.models import KeyedVectors

model = Word2Vec(sentences=review_df.comment_n, window=5, min_count=1, workers=4, sg=0)

# sg=0 -> CBOW, sg=1 -> Skip-gram
```

```python
# ÏôÑÏÑ±Îêú ÏûÑÎ≤†Îî© Îß§Ìä∏Î¶≠Ïä§Ïùò ÌÅ¨Í∏∞ ÌôïÏù∏
model.wv.vectors.shape
```

```python
model_result = model.wv.most_similar("lg")
print(model_result)
```
